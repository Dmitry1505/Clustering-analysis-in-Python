```python
from tkinter import *
from math import sqrt

def solver(a,b,c,x,y,z):
    """ Solves quadratic equation and returns the result in formatted string """
    D = 1.34+0.25*a+0.17*b+0.11*c+0.34*x+0.09*y-0.0003*z
    if D > 3:
        text = "Прогнозируется высокий риск развития СПИД"
    if 2<=D<=3:
        text = "Прогнозируется средний риск развития СПИД"
    if 2>=D:
        text = "Прогнозируется низкий риск развития СПИД" 
    return text

def inserter(value):
    """ Inserts specified value into text widget """
    output.delete("0.0","end")
    output.insert("0.0",value)    

def clear(event):
    """ Clears entry form """
    caller = event.widget
    caller.delete("0", "end")

def handler():
    """ Get the content of entries and passes result to the text """
    try:
        # make sure that we entered correct values
        a_val = float(a.get())
        b_val = float(b.get())
        c_val = float(c.get())
        x_val = float(c.get())
        y_val = float(c.get())
        z_val = float(c.get())
        inserter(solver(a_val, b_val, c_val, x_val, y_val, z_val))
    except ValueError:
        inserter("Введите корректные значения (баллы)")

root = Tk()
root.title("Quadratic calculator")
root.minsize(325,230)
root.resizable(width=False, height=False)


frame = Frame(root)
frame.grid()

a = Entry(frame, width=3)
a.grid(row=3,column=1,padx=(10,0))
a.bind("<FocusIn>", clear)
a_lab = Label(frame, bg="snow",  text="диагноз стадии ВИЧ, баллы").grid(row=3,column=2)

b = Entry(frame, width=3)
b.bind("<FocusIn>", clear)
b.grid(row=3,column=3)
b_lab = Label(frame, bg="snow", text="стадия туберкулеза, баллы").grid(row=3, column=4)

c = Entry(frame, width=3)
c.bind("<FocusIn>", clear)
c.grid(row=3, column=5)
c_lab = Label(frame, bg="snow", text="температура тела, баллы").grid(row=3, column=6)

x = Entry(frame, width=3)
x.grid(row=5,column=1,padx=(10,0))
x.bind("<FocusIn>", clear)
x_lab = Label(frame, bg="snow", text="обструктивные болезни легких, баллы").grid(row=5,column=2)

y = Entry(frame, width=3)
y.bind("<FocusIn>", clear)
y.grid(row=5,column=3)
y_lab = Label(frame, bg="snow", text="размер печени по М.Г. Курлову, сантиметры").grid(row=5, column=4)

z = Entry(frame, width=3)
z.bind("<FocusIn>", clear)
z.grid(row=5, column=5)
z_lab = Label(frame, bg="snow", text="СД4, баллы").grid(row=5, column=6)

but = Button(frame, text="Рассчитать будущий риск развития СПИД", bg="orange red", command=handler).grid(row=10, column=2, columnspan=7 )

output = Text(frame, bg="Azure", font="Arial 12", width=83, height=5)
output.grid(row=7, column=1, columnspan=7)

Text = Label (frame,  bg="azure", text="ПРОГНОЗИРОВАНИЕ РИСКА РАЗВИТИЯ СПИДА У ОСУЖДЕННЫХ В ВОЗРАСТЕ ДО 40 ЛЕТ", font="Arial 10" ).grid(row=1, column=1, columnspan=7 )
Text = Label (frame,   text=" ").grid(row=2)
Text = Label (frame,   text=" ").grid(row=6)
Text = Label (frame,   text=" ").grid(row=9)
Text = Label (frame,   text=" ").grid(row=11)

#logo = PhotoImage(file="C:\Users\Dmitry\Desktop\Калькулятор\1234.gif")
#logo1=Label (root, image=logo)
#logo1.pack()

root.mainloop()

```
